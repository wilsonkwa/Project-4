# Data pre-processing

```python
data02tr = data02[['Latitude', 'Longitude', 'Species_CULEX ERRATICUS', 'Species_CULEX PIPIENS',
                   'Species_CULEX PIPIENS/RESTUANS', 'Species_CULEX RESTUANS', 'Species_CULEX SALINARIUS',
                   'Species_CULEX TARSALIS', 'Species_CULEX TERRITANS', 'Species_UNSPECIFIED CULEX', 'DewPoint',
                   'ResultSpeed', 'ResultDir', 'WnvPresent']]
```

```python
dt_t, dt_h = train_test_split(data02tr, test_size=0.3, random_state=10, stratify=data02tr['WnvPresent'])
print(dt_t.shape)
print(dt_h.shape)
```

```python
#keep all event (WnvPrensent = 1) in a data frame for oversampling purpose below.
dt_t1 = dt_t[dt_t.WnvPresent == 1]
dt_t0 = dt_t[dt_t.WnvPresent == 0]
```

```python
dt_t1_copy = dt_t1.copy()
```

```python
#oversampling event to deal with rare event imbalance data.
dt_t01 = pd.concat([dt_t, dt_t1], axis=0)
dt_t01 = pd.concat([dt_t01, dt_t1], axis=0)
dt_t01 = pd.concat([dt_t01, dt_t1], axis=0)
dt_t01 = pd.concat([dt_t01, dt_t1], axis=0)
dt_t01 = pd.concat([dt_t01, dt_t1], axis=0)
dt_t01 = pd.concat([dt_t01, dt_t1], axis=0)
dt_t01 = pd.concat([dt_t01, dt_t1], axis=0)#repeated till event (Wnv = 1) ~3k (~30%)
```

```python
dt_t01.shape
```

```python
dt_t01 = dt_t01.sample(frac=1)
```

```python
dt_t01.head()
```

```python
X_t1 = dt_t01.drop('WnvPresent', axis=1)
y_t1 = dt_t01['WnvPresent']

X_h1 = dt_h.drop('WnvPresent', axis=1)
y_h1 = dt_h['WnvPresent']

print('X train: ', X_t1.shape)
print('y train: ', y_t1.shape)

print('X test: ', X_h1.shape)
print('y test: ', y_h1.shape)
```

# pipeline (start) - Data Modelling

```python
X_train = dt_t01.drop('WnvPresent', axis=1)
y_train = dt_t01['WnvPresent']

X_test = dt_h.drop('WnvPresent', axis=1)
y_test = dt_h['WnvPresent']
```

```python
print('X train: ', X_train.shape)
print('y train: ', y_train.shape)

print('X test: ', X_test.shape)
print('y test: ', y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, BaggingClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedShuffleSplit, GridSearchCV
from sklearn.metrics import accuracy_score, roc_auc_score
```

```python
model = []
parameters = []
scores = []
precision = []
recall = []
roc_auc = []

# To put the name of model, models from sklearn, and hyperparameters of the model we want to search through in order

names = ['Decision Tree', 'RandomForest', 'LogReg', 'Adaboost', 'BaggingClass', 'GradientBoost']

estimators = [DecisionTreeClassifier(), RandomForestClassifier(), LogisticRegression(),
              AdaBoostClassifier(), BaggingClassifier(), GradientBoostingClassifier()]

#Originally, in each method, there are some alternatives for each parameters.
#Run each methods one by one then selected the best parameters to save computing time.
#Below parameters have been filled with the best option.

params = [
    {'clf__max_depth': [None],'clf__max_features':[2],
     'clf__min_samples_split':[5]},
    {'clf__max_depth': [None],'clf__max_features':[2],
     'clf__min_samples_split':[2]},
    {'clf__penalty': ['l1', 'l2']},
    {"clf__learning_rate": [0.2],"clf__n_estimators":[200]},
    {"clf__max_samples": [0.5, 1.0],"clf__max_features": [10,11],
    "clf__bootstrap": [False],"clf__bootstrap_features": [True]},
    {'clf__learning_rate':[1], 'clf__n_estimators':[5000]}
]

for name, estimator, param in zip(names, estimators, params):
    clf_pipe = Pipeline([
        ('ss', StandardScaler()),
        ('clf', estimator)
    ])
####cv = StratifiedKFold(shuffle=True, n_splits=3)
    cv = StratifiedShuffleSplit(n_splits=3, test_size=0.2, random_state=0)
    gs = GridSearchCV(clf_pipe, param_grid=param, cv=cv, verbose=1, n_jobs=2)
    gs.fit(X_train, y_train)
    
    print(name)
    print('Best Score: {}'.format(gs.best_score_))
    print('Best Params: {}'.format(gs.best_params_))
    
    # Best Model
    grid_model = gs.best_estimator_
    
    print('Model Score on X_train: {}'.format(grid_model.score(X_train, y_train)))
    
    # Prediction and score
    pred = grid_model.predict(X_test)
    print('Accuracy score of prediction: {}'.format(accuracy_score(y_test, pred)))
    print ('ROC_AUC score: {}'.format(roc_auc_score(y_test,pred)))
    
    # Confusion Matrix 
    conf_matrix = confusion_matrix(y_test, pred)
    
    tn, fp, fn, tp = confusion_matrix(y_test, pred).ravel()
    print('TN: {}, FP:{}, FN:{}, TP:{}'.format(tn,fp,fn,tp))
    print('Precision of the model is {}'. format(tp/(tp+fp)))
    print('Recall of the model is {}'.format(tp/(tp+fn)))
    
    print('\n')
    model.append(name)
    parameters.append(gs.best_params_)
    scores.append(accuracy_score(y_test, pred))
    recall.append(tp/(tp+fn))
    precision.append(tp/(tp+fp))
    roc_auc.append(roc_auc_score(y_test,pred))
    
pd.DataFrame({'model': model, 'parameters': parameters, 'scores': scores, 
              'recall': recall, 'precision': precision, 'roc_auc':roc_auc})
```
